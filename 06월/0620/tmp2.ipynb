```python
import pandas as pd
import numpy as np

# 컬럼 제한은 10개로, 시리즈 값은 모두 출력
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_seq_item', None)

# 예시 데이터프레임 생성 (실제 데이터프레임 df로 대체)
np.random.seed(0)
df = pd.DataFrame(np.random.randint(0, 100, size=(8378, 123)), columns=[f'col_{i}' for i in range(123)])

# 각 컬럼별 유니크 값 개수 계산
unique_counts = df.nunique()

# 123개 값이 모두 출력됨
print(unique_counts)
```
