```python
import os
import pandas as pd
import zipfile
```

```python
path = "./data_work/"
with zipfile.ZipFile("./data/2016.zip") as f:
    f.extractall(path)
```

```python
dfs = {}
n = 0
for file in os.listdir(path):
    if file.split(".")[-1] =="csv":
        dfs[n] = pd.read_csv(path + file, encoding='euc-kr')
        n += 1
len(dfs)
```

```python
for i in range(12):
    print(dfs[i].columns)
```

```python
for i in range(12):
    dfs[i].columns = dfs[i].columns.str.replace(' ', '').str.replace('~', '-').str.replace('시', '')
```

```python
for i in range(12):
    print(dfs[i].columns)
```

```python
new_cols = []
for i in range(12):
    for col in dfs[i]:
        if not col in new_cols:
            new_cols.append(col)
```

```python
new_cols
```

```python
times = new_cols[5:]
times.sort()
new_cols = new_cols[0:5] + times
```

```python
new_cols
```

```python
total = pd.DataFrame(columns=new_cols)
for i in range(12):
    total = pd.concat([total, dfs[i]])
```
