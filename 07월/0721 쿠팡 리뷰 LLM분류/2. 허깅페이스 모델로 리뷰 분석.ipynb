```python
import re

def clean_str(lst):
    for idx in range(len(lst)):
        lst[idx] = lst[idx].replace('\n', '').replace('\r', '')
        pattern = r"[^ㄱ-ㅎㅏ-ㅣ가-힣A-Za-z0-9,\. ]"
        lst[idx] = re.sub(pattern, "", lst[idx])
```

```python
clean_str(milk_list)
clean_str(egg_list)
clean_str(water_list)
```
