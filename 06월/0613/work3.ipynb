```python
import os
import pandas as pd
import zipfile

path = "./data_work/"
with zipfile.ZipFile("./data/2016.zip") as f:
    f.extractall(path)
```

```python
dfs = [pd.read_csv(path + file, encoding='euc-kr') for file in os.listdir(path) if file.split(".")[-1] =="csv" ]
```

```python
[print(df.columns) for df in dfs]
```

```python
for df in dfs:
    df.columns = df.columns.str.replace(' ', '').str.replace('~', '-').str.replace('시', '')
```
