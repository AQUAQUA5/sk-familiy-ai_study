```python
cu_url = "https://www.pocketcu.co.kr/api/store/search/list"

cnt = 20
n=0
while cnt == 20:
    cu_payload = {'firstRowNum' : f"{n}",
        'searchWord' : ""}
    res = requests.post(cu_url, json=cu_payload).json()['storeList']
    cnt = len(res)
    print(n, cnt)
    n +=1
print((n)*20 + cnt)
```
