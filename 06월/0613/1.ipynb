```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.formula.api as smf
```

```python
%precision 3
%matplotlib inline
```

```python
df = pd.read_csv('./data/scores_reg.csv')
n = len(df)
print(n)
df.head()
```

```python
df.columns = ['quiz',   'final_test',   'sleep_time',   'school_method']
df.loc[df.school_method == 'バス', 'school_method'] = 'bus'
df.loc[df.school_method == '自転車', 'school_method'] = 'bicyle'
df.loc[df.school_method == '徒歩', 'school_method'] = 'walk'
```

```python
df.head(3)
```

```python
x = np.array(df['quiz'])
y = np.array(df['final_test'])
p = 1
```

```python
poly_fit = np.polyfit(x, y, 1)
poly_1d = np.poly1d(poly_fit)
xs = np.linspace(x.min(), x.max())
ys = poly_1d(xs)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.set_xlabel('quiz')
ax.set_ylabel('final test')
ax.plot(xs, ys, color='gray', 
        label=f'{poly_fit[1]:.2f}+{poly_fit[0]:.2f}x')
ax.scatter(x, y)
ax.legend()

plt.show()
```

```python
import statsmodels.formula.api as smf
formula = 'final_test ~ quiz'
result = smf.ols(formula, df).fit()
result.summary()
```
